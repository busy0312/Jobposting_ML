```python
from pyspark.sql import SparkSession
from pyspark import SparkFiles
import pyspark.sql.functions as F
from pyspark.ml.feature import (
   Tokenizer, 
    StopWordsRemover, 
    HashingTF, 
    IDF, 
    StringIndexer
)
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.evaluation import MulticlassClassificationEvaluator
from pyspark.ml.classification import NaiveBayes
```

```python
# get or create Spark session

app_name = "job-john-snow"
spark = SparkSession.builder.appName(app_name).getOrCreate()
```

```python
file_uri = "/mnt/mnt_project3/job.csv"

df = spark.read \
    .format("com.databricks.spark.csv") \
    .options(header='true', inferSchema="true") \
    .load(file_uri)
df.printSchema()
```

```python
data=df.select('description','fraudulent')
df_new=data.filter(data.description.isNotNull())
df_new=df_new.filter(data.fraudulent.isNotNull())
data_df=df_new.filter(data.fraudulent<2)
data_df.show()
```

```python
from pyspark.sql.functions import isnan, when, count, col
data_df.select([count(when(col(c).isNull(), c)).alias(c) for c in data.columns]).show()
data_df.count()
```

```python
display(
data_df.filter(data_df.fraudulent==1)
)
```

```python
import sparknlp
from sparknlp.base import *
from sparknlp.annotator import *
from pyspark.ml import Pipeline
from sparknlp.pretrained import PretrainedPipeline
```

```python
document = DocumentAssembler()\
    .setInputCol("description")\
    .setOutputCol("document")

use = UniversalSentenceEncoder.pretrained() \
 .setInputCols(["document"])\
 .setOutputCol("sentence_embeddings")

# the classes/labels/categories are in category column
classsifierdl = ClassifierDLApproach()\
  .setInputCols(["sentence_embeddings"])\
  .setOutputCol("class")\
  .setLabelColumn("fraudulent")\
  .setMaxEpochs(5)\
  .setEnableOutputLogs(True)

pipeline = Pipeline(
    stages = [
        document,
        use,
        classsifierdl
    ])
```

```python
pipelineModel = pipeline.fit(data_df)
```

```python
#save it to s3
pipelineModel.save("/mnt/mnt_s3/job_pipeline")
```

```python
results = pipelineModel.transform(data_df)
display(
  results
)
```

```python
display(
  results.selectExpr("description",'fraudulent','class.result as prediction')
)
```

```python
pdf = results.selectExpr("fraudulent", "class.result").toPandas()
pdf.head()
```

```python
from sklearn.metrics import classification_report
print(classification_report(pdf.fraudulent.astype(float), pdf.result.astype(float)))
```
